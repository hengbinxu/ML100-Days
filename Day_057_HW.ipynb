### 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering

np.random.seed(123)

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
estimator = [('hc_ward', AgglomerativeClustering(n_clusters=3, linkage='ward')),
             ('hc_average', AgglomerativeClustering(n_clusters=3, linkage='average')),
             ('hc_complete', AgglomerativeClustering(n_clusters=3, linkage='complete'))]
```

```python
fig_num = 1
for name, est in estimator:
    fig = plt.figure(num=fig_num, figsize=(4, 3))
    ax = Axes3D(fig, rect=[0, 0, 1, 1], elev=48, azim=134)
    label_ = est.fit_predict(X)
    
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=label_)
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(name)
    ax.dist = 12
    
    fig_num += 1
    plt.show()
    
# plot the ground truth
fig = plt.figure(figsize=(4, 3))
num_cluster = [('cls_0', 0), ('cls_1', 1), ('cls_2', 2)]
ax = Axes3D(fig, rect=[0, 0, 1, 1], elev=48, azim=134)
for name, num_c in num_cluster:
    ax.text3D(X[y==num_c, 0].mean(), 
              X[y==num_c, 1].mean(),
              X[y==num_c, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))  
    
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, edgecolor='k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_title('The ground truth')
ax.dist = 12
plt.show()
```
