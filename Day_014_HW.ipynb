### 請根據不同的 HOUSETYPE_MODE 對 AMT_CREDIT 繪製 Histogram

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
unique_house_type = app_train['HOUSETYPE_MODE'].unique()
```

```python
"""
Your Code Here
"""
unique_house_type = app_train['HOUSETYPE_MODE'].unique()
is_missing = pd.isna(unique_house_type)

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    if is_missing[i] == False:
        house_type = unique_house_type[i]
        _filter = app_train['HOUSETYPE_MODE'] == house_type
    else:
        _filter = pd.isna(app_train['HOUSETYPE_MODE'])
        house_type = 'others'
    app_train.loc[ _filter, 'AMT_CREDIT'].hist()
    plt.title(str(house_type))
plt.show()    
```
