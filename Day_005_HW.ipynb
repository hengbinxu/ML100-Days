```python
def calculate_column_mean_std(data, num_col):
    float_colnames = data.select_dtypes(include=['float64']).columns
    colnames = np.random.choice(float_colnames, num_col, replace=False) # random select columns
    sub_data = data[colnames] # get subdata
    average = sub_data[colnames].mean()
    standard_deviation = sub_data[colnames].mean()
    for colname, avg, stand in zip(average.index, average.values, standard_deviation.values):
        print('column: {}, mean: {}, std: {}'.format(colname, avg, stand))
        
    # plot histgram
    fig = sub_data.hist(bins=50, grid=True, layout=(num_col, 2), figsize=(10, 12))
    [x.title.set_size(8) for x in fig.ravel()] # adjust subplots title size
    plt.tight_layout() # will also adjust spacing between subplots to minimize the overlaps
    plt.show()
```

```python
calculate_column_mean_std(app_train, 6)
```
